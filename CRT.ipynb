```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel('mod2.xlsx', skiprows=1)

# Извлечение признаков и целевой переменной
features = data.iloc[:, 1:7]  # Шесть параметров начиная со второго столбца
target = data.iloc[:, 8]
feature_names = [str(column) for column in features.columns]
class_names = [str(cls) for cls in target.unique()]
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
test_countries = data.iloc[X_test.index, 0]   # Первый столбец - название страны
for country, true_class, predicted_class in zip(test_countries, y_test, predictions):
    print("Country:", country, "- True Class:", true_class, "- Predicted Class:", predicted_class)
```

```python
plt.figure(figsize=(12, 8))
plt.axis('off')  # Отключение осей координат
plt.title('Decision Tree', fontsize=16)
plt.annotate('',
             xy=(0.5, 0.95),
             xytext=(0.5, 0.90),
             arrowprops=dict(arrowstyle='->', lw=1.5))
plt.text(0.5, 0.96, 'Root', ha='center', va='center', fontsize=12, fontweight='bold')
plot_tree(model, feature_names=feature_names, class_names=class_names, filled=True)
plt.show()
```
