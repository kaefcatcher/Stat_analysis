```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
Data=pd.read_excel('Semestr2.xlsx', sheet_name='for_python2') 
Data.head() 
```

```python
newData=Data.iloc[:,[1,2,3,4,5,6,7,8]].values
```

```python
from sklearn.cluster import AgglomerativeClustering 
Agg_hc = AgglomerativeClustering(n_clusters = 6, affinity = 'euclidean', linkage = 'ward')
y_hc = Agg_hc.fit_predict(newData) 
```

```python
plt.scatter(newData[y_hc == 0, 0], newData[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1') 
plt.scatter(newData[y_hc == 1, 0], newData[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2') 
plt.scatter(newData[y_hc == 2, 0], newData[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3') 
plt.scatter(newData[y_hc == 3, 0], newData[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')  
plt.scatter(newData[y_hc == 4, 0], newData[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(newData[y_hc == 5, 0], newData[y_hc == 5, 1], s = 100, c = 'yellow', label = 'Cluster 6')

plt.legend()
plt.show()
```

```python
Agg_hc = AgglomerativeClustering(n_clusters = 6, affinity = 'euclidean', linkage = 'complete')
y_hc = Agg_hc.fit_predict(newData) 
plt.scatter(newData[y_hc == 0, 0], newData[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1') 
plt.scatter(newData[y_hc == 1, 0], newData[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2') 
plt.scatter(newData[y_hc == 2, 0], newData[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(newData[y_hc == 3, 0], newData[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')  
plt.scatter(newData[y_hc == 4, 0], newData[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(newData[y_hc == 5, 0], newData[y_hc == 5, 1], s = 100, c = 'yellow', label = 'Cluster 6')

plt.legend()
plt.show()
```
