```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
df=pd.read_excel('Semestr2.xlsx', sheet_name='z-clusters')  
```

```python
df.describe()
```

```python
scaler=StandardScaler()
```

```python
df.columns
```

```python
kmeans= KMeans(n_clusters=6)
```

```python
kmeans.fit(df[['Индекс потребительских цен, 2010 г.=100',
       'ВВП на душу населения в ценах и ППС(США) 2010 г., темп роста']])
```

```python
df['kmeans_6'] = kmeans.labels_
```

```python
df
```

```python
means = df.iloc[:, 2:].groupby('kmeans_6').mean()
means
```

```python
clusters = df.groupby('kmeans_6')
sns.set_style('whitegrid')

# создание графика
fig, ax = plt.subplots(figsize=(10, 6))

# отрисовка точек для каждого кластера
for name, cluster in clusters:
    ax.scatter(cluster['Импорт товаров и услуг, доля в процентах от ВВП'], cluster['ВВП на душу населения в ценах и ППС(США) 2010 г., темп роста'], 
               s=100, alpha=0.8, label=name)

# добавление легенды
ax.legend(title='Кластер')




# отображение графика
plt.show()
```
